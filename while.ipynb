## <kbd style="background:#CCFFCC">2</kbd> *Ausführen:* Das Programm testen

<div class="alert alert-block alert-info" style="color:#000">
<strong>⚙️ Führt diesen Code nun aus.</strong>

Dazu wählt ihr die Code-Zelle unten aus, indem ihr sie anklickt (achtet auf die Markierung links daneben).<br />
Anschließend könnt ihr den Code entweder mit <kbd>STRG</kbd> + <kbd>ENTER</kbd> oder durch Klick auf <kbd>▶</kbd> in der oberen Leiste ausführen.
</div>

````python
import random

number = random.randint(1, 100)
guess = int(input("Errate meine Zahl zwischen 1 und 100:"))

while guess != number:
    if number > guess:
        print("⬆️ Meine Zahl ist größer.")
    elif number < guess:
        print("⬇️ Meine Zahl ist kleiner.")
    guess = int(input("Versuch es noch einmal:"))

print("✅ Richtig! Meine Zahl ist " + str(number) + ".")
````

````python
import random

number = random.randint(1, 100)
guess = int(input("Errate meine Zahl zwischen 1 und 100:"))

while guess != number:
    if number > guess:
        print("⬆️ Meine Zahl ist größer.")
    elif number < guess:
        print("⬇️ Meine Zahl ist kleiner.")
    guess = int(input("Versuch es noch einmal:"))

print("✅ Richtig! Meine Zahl ist " + str(number) + ".")
````
